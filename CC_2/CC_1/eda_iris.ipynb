# Conduct exploratory data analysis on the famous Iris dataset to understand its characteristics and relationships between features.

```python
# imports
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Enable inline plotting
%matplotlib inline

from mpl_toolkits.mplot3d import Axes3D
```

```python
# Load dataset
df = sns.load_dataset('iris')
df.head()
df.info()
df.describe()   # statistics summary
df.isnull().sum()
```

```python
# class distribution
df['species'].value_counts()

sns.countplot(x='species', data=df)
plt.title('Species Count')
plt.show()
```

```python
# numerical features
df.hist(figsize=(10, 8), color='lightblue', edgecolor='black')
plt.tight_layout()
plt.show()

# histogram with KDE curve
for col in df.columns[:-1]:
    sns.histplot(df[col], kde=True, color='darkcyan')
    plt.title(f'Distribution of {col}')
    plt.show()
```

```python
# outliers
for col in df.columns[:-1]:
    sns.boxplot(x='species', y=col, data=df)
    plt.title(f'{col} by Species')
    plt.show()

```

```python
# pair-wise relationships
sns.pairplot(df, hue='species', corner=True, palette='husl')
plt.suptitle("Pairwise Relationships", y=1.02)
plt.show()
```

```python
# Correlation matrix
corr = df.drop('species', axis=1).corr()

sns.heatmap(corr, annot=True, cmap='YlGnBu', linewidths=0.5)
plt.title("Correlation Matrix")
plt.show()
```

```python
# 3D Visualization 

fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')

# Map species to numbers
colors = df['species'].map({'setosa': 0, 'versicolor': 1, 'virginica': 2})

ax.scatter(df['sepal_length'], df['petal_length'], df['petal_width'], c=colors, cmap='viridis')
ax.set_xlabel('Sepal Length')
ax.set_ylabel('Petal Length')
ax.set_zlabel('Petal Width')
plt.title('3D Scatter Plot')
plt.show()
```
